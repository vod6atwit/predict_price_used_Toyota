# Data Preprocessing

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
dataset = pd.read_csv('./dataSet/cars_dataset.csv')
dataset.head(100)
```

## Create new Data Set 

#### Get only "toyota" cars 

```python
new_dataset = dataset[dataset['Make'] == "toyota"].reset_index()
new_dataset
```

```python
new_dataset = new_dataset.drop(columns=['index', 'Make'])
new_dataset
```
